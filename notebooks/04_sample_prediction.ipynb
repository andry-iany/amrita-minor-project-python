# Prediction

```python
import numpy as np

from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.text import tokenizer_from_json
from keras.preprocessing.sequence import pad_sequences


from sentiment_analysis.utils import PATH_MODEL, PATH_TOKENIZER, clean_text
```

```python
# Load the entire model
model = load_model(PATH_MODEL)

# Verify the structure
model.summary()
```

```python
# load the tokenizer
with open(PATH_TOKENIZER, 'r', encoding="utf-8") as f:
    data = f.read()
    tokenizer = tokenizer_from_json(data)
```

```python
tests = [
    "What a wonderful movie", 
    "This was a terrible movie",
]

X_test = np.array(list(map(clean_text, tests)))
```

```python
# Padding all reviews to fixed length 250
maxlen = 250
padding_type = 'post'

# tokenize and pad the test data
X_test = tokenizer.texts_to_sequences(X_test)
X_test = pad_sequences(X_test, padding=padding_type, maxlen=maxlen)
```

```python
model.predict(X_test)
```
